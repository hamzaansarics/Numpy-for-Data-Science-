```python
from numpy.fft import fft,fftfreq
import numpy as np
import matplotlib.pyplot as plt
Lx=100
n=1000
omg=2.0*np.pi/n
x=np.linspace(1,Lx,n)
y1=1.0*np.cos(5.0*omg*x)
y2=2.0*np.sin(10.0*omg*x)
y3=3.0*np.sin(20.0*omg*x)
y=y1+y2+y3
freq=fftfreq(n)
true=freq>0
fft_val=fft(y)
nwaves=freq*n
fft_theo=2.0*np.abs(fft_val/n)
ps=2.0*(np.abs(fft_val/n)**2.0)
fps=freq*ps
plt.plot(x,y)
plt.show()
plt.plot(fft_val[true],ps[true])
plt.semilogx(freq[true],fps[true])
```
