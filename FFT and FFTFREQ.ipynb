```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft,fftfreq
Lx=100
n=1000
omg=2.0*np.pi/Lx
x=np.linspace(1,Lx,n)
y1=1.0*np.cos(5.0*omg*x)
y2=2.0*np.sin(10.0*omg*x)
y3=3.0*np.sin(20*omg*x)
y=y1+y2+y3
fftfrequencies=fftfreq(n)
true=fftfrequencies>0
fft_values=fft(y)
fft_theo=2.0*np.abs(fft_values/n)
plt.figure(1)
plt.plot(x,y,label='Orignal Signal',color='xkcd:salmon')
plt.legend()
plt.show()
# plt.figure(2)
# plt.plot(fftfrequencies,fft_values,label='RAW FFT ',color='c')
# plt.legend()
# plt.show()
# plt.figure(3)
# plt.plot(fftfrequencies[true],fft_theo[true],label='More Processed RAW FFT',color='g')
# plt.show()
```
