```python
import random
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')
xs,ys=[],[]
fig=plt.figure()
for i in range(10):
    x=i
    y=random.random()
    xs.append(x)
    ys.append(y)
ax=fig.add_subplot(221)
ax.plot(xs,ys)
ax=fig.add_subplot(222)
ax.plot(xs,ys)
ax=fig.add_subplot(212)
ax.plot(xs,ys)
plt.show()
ax=plt.subplot2grid((6,1),(0,0),rowspan=1,colspan=1)
ax.plot(xs,ys)
ax=plt.subplot2grid((6,1),(1,0),rowspan=4,colspan=1)
ax.plot(xs,ys)
ax=plt.subplot2grid((6,1),(5,0),rowspan=1,colspan=1)
ax.plot(xs,ys)
```
