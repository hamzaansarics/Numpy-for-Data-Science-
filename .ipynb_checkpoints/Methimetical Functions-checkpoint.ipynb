```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
img=cv2.imread('D:\\girl.jpg')
ls=np.linspace(1,10,5)
ar=np.arange(-10,10.5,0.1)
def mine(x,mean,var):
    return (1.0/(2.0*np.pi*var)**0.5)*np.exp(-(x-mean)**2)/(2.0*var)
x=np.arange(-5,5.01,0.05)
pdf1=mine(x,2,1)
pdf2=mine(x,1,2)
pdf3=mine(x,1.5,0.6)
pdf4=mine(x,2.5,5.6)
plt.plot(x,pdf1,label='x=0,mean=2,var=1')
plt.plot(x,pdf2,label='x=0,mean=1,var=2')
plt.plot(x,pdf3,label='x=0,mean=1.5,var=0.6')
plt.plot(x,pdf4,label='x=0,mean=2.5,var=5.6')
plt.legend()
plt.show()
```
