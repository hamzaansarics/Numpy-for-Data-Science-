```python
import matplotlib.pyplot as plt
import random
import numpy as np
years=[y for y in range(2000,2019)]
prices=[p*100 if p%2==0 else p*8 for p in range(201,220)]
new_prices=[]
fig=plt.figure()
ax=plt.subplot2grid((1,1),(0,0))
for p in prices:
    rand=random.random()
    if rand>0.2 and rand<0.5:
        new_prices.append(p*20)
    elif rand>0.5 and rand<0.8:
        new_prices.append(p*59)
    elif rand>=0.8 or rand<0.10:
        new_prices.append(p*90)
    else:
        new_prices.append(p*30)
avg=0
for p in new_prices:
    avg+=p
avg=round(avg/len(new_prices))
pr=np.array(new_prices)
ax.fill_between(years,pr,avg,facecolor='g',alpha=0.5)
plt.xlabel('Years')
plt.ylabel('Prices')
ax.xaxis.label.set_color('c')
ax.yaxis.label.set_color('g')
ax.set_xticks([2000,2006,2010,2018])
plt.show()
```
