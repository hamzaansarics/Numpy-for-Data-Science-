```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5]
working=[3,7,6,5,1]  #21
sleeping=[6,7,8,3,9] #33
playing=[3,4,5,2,1]  #15
eating=[1,2,3,4,4]   #14
activites=['working','sleeping','playing','eating']
colors=['r','g','b','c']
slices=[21,33,15,14]
plt.pie(slices,
        labels=activites,
        colors=colors,
        startangle=90,
        autopct='%1.1f%%',
        shadow=True,
        explode=(0,0.1,0,0)
       )
plt.show()
```
